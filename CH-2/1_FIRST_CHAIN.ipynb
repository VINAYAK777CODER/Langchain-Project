# **Chains in Langchain**

```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

```

# *First chain*

```python
# Task 1 - Prompt 

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# Task 2- LLM

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
# Task 3 -str parser

str_parser=StrOutputParser()
```

### **manual invocation**

```python
# ✅ Step-by-step execution
template = prompt_template.invoke({"input": "What is the capital of France?"})

res = llm.invoke(template)

final_result = res.content

print(final_result)
```

### **chain invocation**

```python
# ✅ Chain
chain = prompt_template | llm | str_parser

# ✅ Invoke
result = chain.invoke({"input": "What is the capital of France?"})

print(result)
```
