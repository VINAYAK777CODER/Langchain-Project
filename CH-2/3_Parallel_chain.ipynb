# **PARALLEL CHAIN**

```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
# Task 1 - [Prompt] 

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# Task 2- [LLM]

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
# Task 3 -[str parser]

str_parser=StrOutputParser()
```

```python
# Task 4 - [custom runnable]

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text:str)->dict:
    return {"text":text}

dictionary_maker_runnable=RunnableLambda(dictionary_maker)
```

## **Parallel chain -1**

```python
# TASK - 1 [Template For linkedin Post]
linkedin_post = ChatPromptTemplate.from_messages(
    messages=[
        ("system", "You're a LinkedIn post generator. Output ONLY the post text. No intros like 'Here is a post', no headers, no meta-commentary. Start directly with the post content and Be concise, max 100 words."),
        ("human", "{text}"),
    ]
)

# Task 2- [LLM]

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

# Task 3 -[str parser]

str_parser=StrOutputParser()

chain_linkedin= linkedin_post | llm | str_parser
```

## **parallel chain-2**

```python
from langchain_core.runnables import RunnableParallel
```

```python
def insta_chain(text: dict):

    text = text["text"]

    # TASK - 1 [Prompt]
    insta_prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an Instagram post generator. Output ONLY the post. Be concise, max 100 words"),
        ("human", "Create a post for the following text for Instagram: {text}")
    ])

    # TASK - 2 [LLM]
    llm = ChatGroq(
        model="llama-3.3-70b-versatile",
        temperature=0
    )

    # TASK - 3 [Str Parser]
    str_parser = StrOutputParser()

    chain_insta = insta_prompt | llm | str_parser

    result = chain_insta.invoke({"text": text})
    # print(result)

    return result
    

insta_chain_runnable=RunnableLambda(insta_chain)
```

## **Final orchestration**

```python
final_chain = (
    prompt_template |
    llm |
    str_parser |
    dictionary_maker_runnable |
    RunnableParallel(linkedin=chain_linkedin, instagram=insta_chain_runnable)
)
```

```python
topic = input("Enter a topic: ")
final_chain.invoke({"input": topic})
```
